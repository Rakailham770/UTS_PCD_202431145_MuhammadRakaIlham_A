```python
img = cv2.imread("RAKA.jpeg")
```

```python
plt.imshow(img)
plt.axis('off') 
plt.show()
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
plt.axis('off') 
plt.show()
```

```python
print(img.shape)
```

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray",gray)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
print(gray.shape)
```

```python
fig, axs = plt.subplots(2, 2, figsize = (15, 5))

# Tambahkan cmap='gray' di sini
axs[0,0].imshow(gray, cmap='gray') 
axs[0,1].hist(gray.ravel(), 256, [0, 256])

hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
# Dan tambahkan juga di sini
axs[1,0].imshow(gray, cmap='gray')
axs[1,1].plot(hist)

plt.show()
```

```python
beta = 20
baris, kolom = gray.shape 

# Buat img_cerah dengan 2 dimensi saja (karena gray itu 2D)
img_cerah = np.zeros((baris, kolom), dtype=np.uint8)

for x in range(baris):
    for y in range(kolom):
        # Gunakan int() agar tidak overflow saat proses tambah
        gyx = int(gray[x, y]) + beta
        # Batasi agar tidak lewat dari 255
        img_cerah[x, y] = min(255, gyx)
        
img_cerah = img_cerah.astype(np.uint8)

fig, axs = plt.subplots(2, 2, figsize = (15, 5))

axs[0,0].imshow(gray, cmap='gray')
axs[0,1].hist(gray.ravel(), 256, [0, 256])
axs[1,0].imshow(img_cerah, cmap='gray')
axs[1,1].hist(img_cerah.ravel(), 256, [0, 256])
```

```python
alpha = 1.2
baris, kolom = gray.shape[:2] # Mengambil dimensi dari gambar input

# Inisialisasi dengan nol
gray_kontras = np.zeros((baris, kolom), dtype=np.float32) 

for x in range(baris):
    for y in range(kolom):

        gyx = gray[x, y] * alpha
        gray_kontras[x, y] = np.clip(gyx, 0, 255)


gray_kontras = gray_kontras.astype(np.uint8)

# Visualisasi
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) 
axs[0,0].imshow(gray, cmap='gray') 
axs[0,0].set_title("Original gray")
axs[0,1].hist(gray.ravel(), 256, [0, 256])

axs[1,0].imshow(gray_kontras, cmap='gray')
axs[1,0].set_title(f"Kontras Alpha {alpha}")
axs[1,1].hist(gray_kontras.ravel(), 256, [0, 256])

plt.show()
```

```python
alpha = 1.2  # Nilai kontras
beta = 20    # Nilai kecerahan

gray_float = gray.astype(np.float32)
gray_cerah = np.clip(gray_float + beta, 0, 255).astype(np.uint8)
gray_kontras = np.clip(gray_float * alpha, 0, 255).astype(np.uint8)
gray_combo = np.clip((gray_float * alpha) + beta, 0, 255).astype(np.uint8)
fig, axs = plt.subplots(2, 4, figsize=(20, 8))

axs[0,0].imshow(gray, cmap='gray')
axs[0,0].set_title("1. Original Gray")
axs[1,0].hist(gray.ravel(), bins=256, range=[0, 256])

axs[0,1].imshow(gray_cerah, cmap='gray')
axs[0,1].set_title(f"2. Cerah (Beta={beta})")
axs[1,1].hist(gray_cerah.ravel(), bins=256, range=[0, 256])

axs[0,2].imshow(gray_kontras, cmap='gray')
axs[0,2].set_title(f"3. Kontras (Alpha={alpha})")
axs[1,2].hist(gray_kontras.ravel(), bins=256, range=[0, 256])

axs[0,3].imshow(gray_combo, cmap='gray')
axs[0,3].set_title(f"4. Combo (Alpha={alpha}, Beta={beta})")
axs[1,3].hist(gray_combo.ravel(), bins=256, range=[0, 256])

plt.tight_layout()
plt.show()
```
